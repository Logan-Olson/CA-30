```python
# The Current updated state object
class VolumeControlFSM:
  current_state = None
  LOW = "LOW"
  MEDIUM = "MEDIUM"
  HIGH = "HIGH"

class VolumeDirection:
  UP = "UP"
  DOWN = "DOWN"

# DECLARE the function volume_control_finite_state_machine()
def volume_control_finite_state_machine(transition_state_list, volume_direction):


  if VolumeControlFSM.current_state is None:
    VolumeControlFSM.current_state = transition_state_list[0]
    return
  
  current_index = transition_state_list.index(VolumeControlFSM.current_state)

  if volume_direction == VolumeDirection.UP:
    VolumeControlFSM.current_state = transition_state_list[min(current_index + 1, len(transition_state_list) - 1)]

  elif volume_direction == VolumeDirection.DOWN:
    VolumeControlFSM.current_state = transition_state_list[max(current_index - 1, 0)]


### TEST CASES ###

transition_state_list = [
    VolumeControlFSM.LOW,
    VolumeControlFSM.MEDIUM,
    VolumeControlFSM.HIGH,
]
volume_direction_list = [
    VolumeDirection.UP,
    VolumeDirection.UP,
    VolumeDirection.UP,
    VolumeDirection.UP,
    VolumeDirection.DOWN,
    VolumeDirection.DOWN,
    VolumeDirection.DOWN,
]

for volume_direction in volume_direction_list:
  volume_control_finite_state_machine(
      transition_state_list=transition_state_list,
      volume_direction=volume_direction
  )
  print("Volume Control FSM Current State: {0}\n".format(VolumeControlFSM.current_state))
```
